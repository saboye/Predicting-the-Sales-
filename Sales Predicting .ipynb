```python
X = data['Youtube_Ads']
y = data['Smartphone_Sales']
```

```python
## splitting the data sets into train and test data sets in the ratio of 70:30
from sklearn.model_selection import train_test_split  
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size = 0.7, test_size = 0.3, random_state = 100)
```

```python
X_train.describe()
```

```python
 X_test.describe()
```

```python
# Add a constant to get an intercept
X_train_sm = sm.add_constant(X_train)
```

```python
## using ordinary least square method
lrm = sm.OLS(y_train, X_train_sm).fit()
```

```python
lrm.params
```

```python
print(lrm.summary())
```

```python
# y = mx + c
plt.scatter(X_train,y_train)
plt.plot(X_train , 62.0100+5.45*X_train,'r') 
plt.show()
```

```python
y_train_p=lrm.predict(X_train_sm)
result = y_train-y_train_p
print(result)
```

```python
sns.distplot(result,bins=5)
plt.show()
```

```python
plt.scatter(X_train,result)
plt.show()
```

```python
X_test_sm=sm.add_constant(X_test)
y_p=lrm.predict(X_test_sm)
y_p.head()
```

```python
plt.scatter(X_test,y_test)
plt.plot(X_test,y_p,'r')
plt.show()
```
